### Now if you try to call the read method again, you will see that nothing will be printed on the console:
This is because once you call the read method, the cursor is moved to the end of the text. Therefore, when you call read again, nothing is displayed since there is no more text to print.

A solution to this problem is that after calling the `read()` method, call the `seek()` method and pass 0 as the argument. This will move the cursor back to the start of the text file. Look at the following script to see how this works:

A solution to this problem is that after calling the `read()` method, call the `seek()` method and pass 0 as the argument. This will move the cursor back to the start of the text file. Look at the following script to see how this works:

```python
text_file.seek(0)
print(text_file.read())
```

In many cases this makes the text easier to work with. For example, we can now easily iterate through each line and print the first word in the line.

```python
text_file.seek(0)
for lines in text_file:
  print(lines.split())
```

```python
#Import required libraries :
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize


import nltk
nltk.download("popular")
```

```python
#Tokenize the text by sentences :
sentences = sent_tokenize(raw2)

#How many sentences are there? :
print (len(sentences))

#Print the sentences :
#print(sentences)
sentences
```

```python
#Tokenize the text with words :
words = word_tokenize(raw2)

#How many words are there? :
print (len(words))
print("\n")

#Print words :
print (words)
words[:10]
```

```python
#Import required libraries :
from nltk.probability import FreqDist

#Find the frequency :
fdist = FreqDist(words)

#Print 10 most common words :
fdist.most_common(10)
```

```python
#Plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
#Empty list to store words:
words_no_punc = []

#Removing punctuation marks :
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the words without punctution marks :
print (words_no_punc)

print ("\n")

#Length :
print (len(words_no_punc))
```

```python
#Frequency distribution :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
#Plot the most common words on grpah:

fdist.plot(10)
```

```python
from nltk.corpus import stopwords

#List of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
#Empty list to store clean words :
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(clean_words)
print("\n")
print(len(clean_words))
```

```python
#Frequency distribution :
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
#Plot the most common words on grpah:

fdist.plot(10)
```
